```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import mnist_reader
from sklearn.metrics import silhouette_score
```

```python

data, y_train = mnist_reader.load_mnist('data/fashion', kind='train')
data = data[:10000]/255 # 1 prosent


N = 15
score = np.zeros(N-3)
var = np.zeros(N-3)

for i in range(3,N):
    print(i-3)
    kmeans = KMeans(n_clusters=i, random_state=0).fit(data)
    var[i-3] = kmeans.inertia_  # Variance
    
    label = kmeans.predict(data)
    score[i-3] = silhouette_score(data, label)
    
    
plt.plot(np.arange(3,N), var)
plt.figure()
plt.plot(np.arange(3,N), score)
plt.figure()

```
